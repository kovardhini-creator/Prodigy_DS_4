```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Sample social media (Twitter-like) data
data = {
    'tweet': [
        "I love the new iPhone update! It's fantastic.",
        "This airline is the worst. Terrible service.",
        "Had a decent experience at the restaurant.",
        "The customer support was super helpful.",
        "I'm never buying from this brand again.",
        "Wow! Amazing performance by the team!",
        "It was okay, not great but not bad either.",
        "Disappointed with the product quality.",
        "Great job on the recent release, loved it!",
        "Such a waste of money, very poor design."
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Function to get sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Apply sentiment analysis
df['sentiment_score'] = df['tweet'].apply(get_sentiment)

# Classify sentiment
df['sentiment_type'] = df['sentiment_score'].apply(
    lambda x: 'Positive' if x > 0.1 else ('Negative' if x < -0.1 else 'Neutral')
)

# Print the processed DataFrame
print(df)

# Pie chart visualization
sentiment_counts = df['sentiment_type'].value_counts()
colors = ['lightgreen', 'lightcoral', 'lightblue']
plt.figure(figsize=(6, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Sentiment Distribution (Pie Chart)')
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.
plt.show()
```
